```python
bang_weather = pd.read_csv("bangalore.csv")
```

```python
precp = bang_weather.loc[:,['date_time','precipMM']]
```

```python
precp['date'] = precp['date_time'].apply(lambda x: x.split()[0])
```

```python
precp['time'] = precp['date_time'].map(lambda x: x.split()[1])
```

```python
precp.head()
```

```python
precp['time_formatted'] = precp['time'].map(lambda x: datetime.strptime(x,"%H:%M:%S").hour)
```

```python
precp.tail()
```

```python
precp['time_formatted'] = precp['time_formatted'].astype('int')
```

```python
precp.dtypes
```

```python
avg_no_of_days = []
for hrs in precp.time_formatted.unique():
    avg_no_of_days.append((precp[precp.time_formatted.eq(hrs) & precp.precipMM.gt(1.0)].shape)[0]/10)
```

```python
y_range = range(0,50,5)
```

```python
plt.plot(avg_no_of_days, marker ='.')
plt.xlabel('Time of the day (hr)')
plt.ylabel('Number of days')
plt.title('Rainfall in different time of day (Average over 10 yrs)') 
plt.xticks(precp.time_formatted.unique())
plt.yticks(y_range)
```
